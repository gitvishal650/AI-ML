```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('MOCK_DATA.csv')
```

```python
df
```

```python
result = df.groupby(['RollNo','Name']).size()
result
```

```python
result = df.groupby(['RollNo','Name'])['fine'].sum()
result
```

```python
result = df.groupby(['Book_title','RollNo'])['Name'].size()
result
```

```python
result = df.groupby(['RollNo','Name']).agg({'fine':['mean','max','min']})
result
```

```python
result = df.groupby('Branch').agg({'fine':'sum'})
result
```

```python
result = df.groupby('issuedate').size()
result
```
